```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style
sns.set(style='whitegrid')

# Load the encoded dataset
df = pd.read_csv('../data/raw/Airbnb_Open_Data.csv')
```

```python
# Convert 'price' column to numeric by removing '$' and converting to float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# Convert 'service fee' column to numeric by removing '$' and converting to float
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)

# Group by 'neighbourhood group' and calculate average price
avg_price_by_neighborhood = df.groupby('neighbourhood group')['price'].mean().reset_index()

# Plot the average price by neighborhood group
plt.figure(figsize=(10, 6))
sns.barplot(x='neighbourhood group', y='price', data=avg_price_by_neighborhood)
plt.title('Average Price by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'neighbourhood group' and calculate average minimum nights
avg_min_nights_by_neighborhood = df.groupby('neighbourhood group')['minimum nights'].mean().reset_index()

# Plot the average minimum nights by neighborhood group
plt.figure(figsize=(10, 6))
sns.barplot(x='neighbourhood group', y='minimum nights', data=avg_min_nights_by_neighborhood)
plt.title('Average Minimum Nights per Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Minimum Nights')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'room type' and calculate average service fee
avg_service_fee_by_room_type = df.groupby('room type')['service fee'].mean().reset_index()

# Plot the average service fee by room type
plt.figure(figsize=(10, 6))
sns.barplot(x='room type', y='service fee', data=avg_service_fee_by_room_type)
plt.title('Average Service Fee by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Service Fee')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of reviews per month
plt.figure(figsize=(10, 6))
sns.histplot(df['reviews per month'].dropna(), bins=50, kde=True)
plt.title('Distribution of Reviews per Month')
plt.xlabel('Reviews per Month')
plt.ylabel('Frequency')
plt.xlim(0, 12)  # Limit x-axis to 12 months
plt.show()
```

```python
# Group by 'neighbourhood group' and calculate average availability
avg_availability_by_neighborhood = df.groupby('neighbourhood group')['availability 365'].mean().reset_index()

# Plot the average availability by neighborhood group
plt.figure(figsize=(10, 6))
sns.barplot(x='neighbourhood group', y='availability 365', data=avg_availability_by_neighborhood)
plt.title('Average Availability (Days) by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Availability (Days)')
plt.xticks(rotation=45)
plt.show()
```
