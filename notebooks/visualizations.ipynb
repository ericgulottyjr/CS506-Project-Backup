```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set plot style
sns.set(style='whitegrid')

# Load the encoded dataset
df = pd.read_csv('../data/raw/Airbnb_Open_Data.csv')
```

```python
# Convert 'price' column to numeric by removing '$' and converting to float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# Convert 'service fee' column to numeric by removing '$' and converting to float
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)

# Group by 'neighbourhood group' and calculate average price
avg_price_by_neighborhood = df.groupby('neighbourhood group')['price'].mean().reset_index()

# Plot the average price by neighborhood group
plt.figure(figsize=(12, 8))
sns.barplot(x='neighbourhood group', y='price', data=avg_price_by_neighborhood, palette='Blues')
plt.title('Average Price by Neighborhood Group', fontsize=16)
plt.xlabel('Neighborhood Group', fontsize=14)
plt.ylabel('Average Price', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'neighbourhood group' and calculate average minimum nights
avg_min_nights_by_neighborhood = df.groupby('neighbourhood group')['minimum nights'].mean().reset_index()

# Sort the values for better visualization
avg_min_nights_by_neighborhood = avg_min_nights_by_neighborhood.sort_values(by='minimum nights', ascending=False)

# Plot the average minimum nights by neighborhood group using a horizontal bar plot with error bars
plt.figure(figsize=(12, 8))
sns.barplot(y='neighbourhood group', x='minimum nights', data=avg_min_nights_by_neighborhood, ci='sd', palette='viridis')
plt.title('Average Minimum Nights per Neighborhood Group', fontsize=16)
plt.ylabel('Neighborhood Group', fontsize=14)
plt.xlabel('Average Minimum Nights', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Group by 'room type' and calculate average service fee
avg_service_fee_by_room_type = df.groupby('room type')['service fee'].mean().reset_index()

# Sort the values for better visualization
avg_service_fee_by_room_type = avg_service_fee_by_room_type.sort_values(by='service fee', ascending=False)

# Plot the average service fee by room type using a bar plot with error bars
plt.figure(figsize=(12, 8))
sns.barplot(x='room type', y='service fee', data=avg_service_fee_by_room_type, ci='sd', palette='crest')
plt.title('Average Service Fee by Room Type', fontsize=16)
plt.xlabel('Room Type', fontsize=14)
plt.ylabel('Average Service Fee', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of reviews per month using a histogram
plt.figure(figsize=(12, 8))
sns.histplot(df['reviews per month'].dropna(), bins=20, color='skyblue', edgecolor='black')

# Add a vertical line for the average number of reviews per month
mean_reviews = df['reviews per month'].dropna().mean()
plt.axvline(mean_reviews, color='red', linestyle='--', linewidth=2, label=f'Average: {mean_reviews:.2f}')

# Add titles and labels
plt.title('Distribution of Reviews per Month', fontsize=16)
plt.xlabel('Reviews per Month', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xlim(0, 12)  # Limit x-axis to 12 months
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Group by 'neighbourhood group' and calculate average availability
avg_availability_by_neighborhood = df.groupby('neighbourhood group')['availability 365'].mean().reset_index()

# Plot the average availability by neighborhood group using a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='neighbourhood group', y='availability 365', data=avg_availability_by_neighborhood, palette='viridis')
plt.title('Average Availability (Days) by Neighborhood Group', fontsize=16)
plt.xlabel('Neighborhood Group', fontsize=14)
plt.ylabel('Average Availability (Days)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
